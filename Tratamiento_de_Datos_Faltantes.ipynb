```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep=';')
datos.head(10)
```

```python
datos.isnull()  # Identifica los datos nulos en el dataset con valor True
```

```python
datos.notnull() # Con este metodo los valores no nulos se visualizan como False
```

```python
datos.info()  # Este metodo nos permite tener el resumen del  total de netradas y datos no nulos
```

```python
datos['Valor'].isnull() # Este comando busca el valor general
```

```python
datos[datos['Valor'].isnull()]  # El datos anterior lo usamos como parametro para solo buscar los no nulos
```

```python
A = datos.shape[0]
datos.dropna(subset = ['Valor'], inplace = True)  # se eliminan los datos nulos
B = datos.shape[0]
A - B
```

```python
datos[datos['Valor'].isnull()]  # Se verifica de nuevo, y efectivamente se han borrado los datos nulos
```

```python
# Se evaluan los datos nulos de la columna mantenimiento, y si el datos nulos corresponde
# a departamento se debe eliminar,
```

```python
datos['Mantenimiento'].isnull() # Se filtran datos totales de la columna
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]  # se filtran solo los datos nulos de la columna
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento') # Criterio a eliminar
```

```python
A = datos.shape[0]
datos = datos[~seleccion] # Se procede con la eliminacion segun criterio de seleccion
B = datos.shape[0]
A - B
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0] # Siguen quedando 1302 filas por impuesto
```

```python
datos = datos.fillna({'Mantenimiento':0, 'Impuesto':0}) # Se reemplazan valores nulos por cero en las columnas
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0] # Se verifican de nuevo datos nulos
```

```python
datos[datos['Impuesto'].isnull()].shape[0]  # Se eliminaron los datos nulos en ambas columnas
```

```python
datos.info()  # Se verifica de nuevo el dataset sin datos nulos
```

```python
datos.to_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep=';', index = False)
datos
```
