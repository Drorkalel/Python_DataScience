## Creando nuevas variables

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep=';')
```

```python
datos.head(10)
```

# Valor bruto como nuevo dato

```python
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
```

# Crear dato Valor/mt2

```python
datos['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor m2'].round(2)  # Se redondea valor a 2 decimales
datos.head(10)
```

# Valor Bruto m2 como nuevo dato

```python
datos['Valor Bruto m2'] = ((datos['Valor Bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
casa = ['Casa', 'Casa en Condominio', 'Casa de Villa']
```

```python
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x:'Casa' if x in casa else 'Departamento' ) # a traves de apply lambda se agrupan varios tipos de datos
datos
```

## Excluyendo Variables

```python
datos.head(10)
```

```python
datos_aux = pd.DataFrame(datos[['Tipo Agrupado','Valor m2','Valor Bruto','Valor Bruto m2']])
datos_aux.head(10)
```

```python
del datos_aux['Valor Bruto']  # Metodo 'del' borra las columnas que le indiquemos
datos_aux.head(10)
```

```python
datos_aux.head(10)
```

```python
datos_aux.pop('Valor Bruto m2') #Segundo metodo para eliminar datos de columnas, aqui muestra un dataseries con el dato eliminado
```

```python
datos.head(10)
```

```python
datos.drop(['Valor Bruto','Valor Bruto m2'], axis = 1, inplace = True)  # En este metodo hay que especificar que son columnas con axis=1
datos.head(10)
```

```python
datos.to_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep = ';', index=False)
```
