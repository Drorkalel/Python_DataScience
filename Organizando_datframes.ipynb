## Organizando DataFrames en Pandas

```python
import pandas as pd
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
df.sort_index(inplace = True) # Ordena las filas
df
```

```python
df.sort_index(  axis = 1) # Se ve reflejado el orden de filas y columnas
df
```

```python
df.sort_values(by = 'X', inplace = True)  # Se ordenan valores por orden ascendente columna X
df
```
