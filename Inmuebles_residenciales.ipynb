## Inmuebles Residenciales

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/alquiler.csv', sep=';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates()) # Se genera una lista con datos unicos
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

```python
datos.head(10)
```

```python
seleccion = datos['Tipo'].isin(residencial) # Se compara la columna tipo del dataset con la lista residencial
seleccion
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates()) # Se puede ver que se respeto la lista filtrada
```

```python
datos.shape[0]  # Cantidad de filas antes del filtrado
```

```python
datos_residencial.shape[0]  # Cantidad de filas despues del filtrado
```

```python
datos_residencial.index = range(datos_residencial.shape[0]) # Se actualizan los indices del nuevo listado
datos_residencial
```

## Exportando Datos

```python
datos_residencial.to_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep=';', index = False)
```
