```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/drive/MyDrive/PYTHON DATASCIENCE/Datascience Pandas/Pandas Tratamiento de Datos/Pandas2/data/datos_residencial.csv', sep = ';')
datos.head(10)
```

```python
datos['Valor'].mean()
```

```python
barrios = ['Ate','Barranco','Comas','Lince','El Agustino','San Luis','Callao']
seleccion = datos['Distrito'].isin(barrios) # Selecciona el listado de barrios que creamos
datos = datos[seleccion]
```

```python
datos.head(10)
```

```python
datos['Distrito'].drop_duplicates()
```

```python
grupo_barrio = datos.groupby('Distrito')
```

```python
type(grupo_barrio)
```

```python
grupo_barrio.groups # agrupa en una lista los distritos
```

```python
for barrio, data in grupo_barrio:
  print('{} -> {}'.format(barrio, data.Valor.mean())) # Calcula la media para cada grupo de distrito
```

```python
grupo_barrio['Valor'].mean().round(2) # El mismo valor de la mediana desde otro metodo para una columna
```

```python
grupo_barrio[['Valor','Mantenimiento']].mean().round(2) # Para dos o mas columnas
```

```python
grupo_barrio[['Valor']].describe().round(2)
```

```python
grupo_barrio['Valor'].aggregate(['min','max']).rename(columns={'min': 'minimo', 'max': 'maximo'})# Se cambian nombres columnas
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(10,5))
```

```python
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor Medio del Alquiler por Distrito', {'fontsize': 14})
```
